Task 1: Find Words

```python
import re 
sentence = "I have a cat and a dog." 
match = re.search(r"dog", sentence) 
if match: 
    print("Found:", match.group())   
```

```python
sentence = "I have a cat and a dog." 
match = re.search(r"cat", sentence) 
if match: 
    print("Found:", match.group())   
```

Task 2: Match Numbers

```python
text = "My phone number is 0321-4567890."
match = re.search(r"\d\d", text) 
if match: 
    print("Found digits:", match.group())   
```

```python
text = "My phone number is 0321-4567890."
match = re.search(r"\d{4}-?\d{7}", text) 
if match: 
    print("Found digits:", match.group())  
```

Task 3: Anchors

```python
text = "Python is fun"
match = re.match(r"^Python", text) 
if match: 
    print("Starts with Python") 
match = re.search(r"fun$", text) 
if match: 
    print("Ends with fun") 
```

Task 4: Special Characters

```python
text = "Version 3.14" 
match = re.search(r"\.", text) 
if match: 
    print("Found dot in version")  
```

```python
text = "Version 3.14"
match = re.search(r"\d+\.\d+", text)
if match:
    print("Decimal number found:", match.group())
```

\d+\.\d+ means:
        One or more digits, followed by a dot, followed by one or more digits.

Task 5: Groups

```python
text = "My favorite fruit is a mango." 
match = re.search(r"My favorite (fruit) is a (mango).", text) 
if match: 
    print("Group1:", match.group(1))   
    print("Group2:", match.group(2))   
```

Task 6: Lookaheads

```python
text = "helloworld"
match = re.search(r"hello(?=world)", text)
if match:
    print("Matched:", match.group())
```

```python
text = "helloworld"
match = re.search(r"(?=python)", text)
print("Result:", match)
```

Task 7: Quantifiers

```python
text = "sooooon"
match = re.search( r"o+", text)
if match:
    print("Matched:", match.group())
```

```python
text = "haha"
match = re.search(r"(ha){2}", text)
if match:
    print("Matched:", match.group())
```

Task 8: Either/Or

```python
text = "I like apple pie" 
match = re.search(r"apple|bnana", text) 
if match: 
    print("Found:", match.group())  
```

```python
text = "I like bnana shake" 
match = re.search(r"apple|bnana", text) 
if match: 
    print("Found:", match.group())
```

```python
text = "Contact us at support@gmail.com or info@mydomain.org"
emails = re.findall(r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}", text)
print("Emails found:", emails)
```
