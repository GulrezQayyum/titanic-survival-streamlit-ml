```python
import re
import nltk
from textblob import TextBlob
nltk.download('punkt')
nltk.download('punkt_tab') 
from nltk.tokenize import word_tokenize
```

```python
texts = [
"OMG!! I just won $200!! I&#39;m so gr8!!",
"u r dng btr thn ny autmtc txt nrmlztion prgrm cn do.",
"M goin to d place 2moro n will b bak soon.",
"My Phn # is 123-456-7890."
]
```

```python
def to_lowercase(text):
    return text.lower()
normalized_texts = [to_lowercase(text) for text in texts]
print(normalized_texts)
```

```python
def remove_special_chars(text):
    return re.sub(r'[^a-zA-Z\s]', '', text)
normalized_texts = [remove_special_chars(text) for text in normalized_texts]
print(normalized_texts)
```

```python
from nltk.tokenize import word_tokenize
```

```python
def tokenize(text):
    return word_tokenize(text)
tokenized_texts = [tokenize(text) for text in normalized_texts]
print(tokenized_texts)
```

```python
abbreviations = {
"u": "you",
"r": "are",
"gr8": "great",
"b4": "before",
"pls": "please",
"2moro": "tomorrow"
}
```

```python
def expand_abbreviations(text):

    words = text.split()
    return ' '.join([abbreviations.get(word, word) for word in words])
expanded_texts = [expand_abbreviations(' '.join(tokens)) for tokens in tokenized_texts]
print(expanded_texts)
```

```python
def correct_spelling(text):
    blob = TextBlob(text)
    return str(blob.correct())
corrected_texts = [correct_spelling(text) for text in expanded_texts]
print(corrected_texts)
```

```python
import inflect
p = inflect.engine()
```

```python
def convert_numbers(text):
    words = text.split()
    new_words = []
    for word in words:
        if word.isdigit():
            new_words.append(p.number_to_words(word))
        elif '$' in word:
            # Convert $200 to 'two hundred dollars'
            new_words.append(p.number_to_words(word.strip('$')) + ' dollars')
        else:
            new_words.append(word)
    return ' '.join(new_words)
```

```python
final_texts = [convert_numbers(text) for text in corrected_texts]
print(final_texts)
```
