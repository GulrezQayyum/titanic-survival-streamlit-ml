### 1. Word Tokenization (Using NLTK)

```python
from nltk.tokenize import word_tokenize

sentence = "I love reading books!"

tokens = word_tokenize(sentence)

print("Word Tokens:", tokens)
```

--- 

### 2. Sentence Tokenization (Using NLTK)

```python
from nltk.tokenize import sent_tokenize

text = "I love reading. It's fun and relaxing!"

sentences = sent_tokenize(text)

print("Sentence Tokens:", sentences)
```

--- 

### 3. Whitespace Tokenization

```python
text = "I love Natural Language Processing"

tokens = text.split(" ")

print("Whitespace Tokens:", tokens)
```

--- 

### 4. Character Tokenization

```python
text = "cat"

tokens = list(text)

print("Character Tokens:", tokens)
```

--- 

### 5. Rule-Based Tokenization (Using spaCy)

```python
import spacy

# Load the English language model

nlp = spacy.load("en_core_web_sm")

doc = nlp("I'm learning NLP with spaCy!")

tokens = [token.text for token in doc]

print("Rule-Based Tokens:", tokens)
```

--- 

### 6. Subword Tokenization (Hugging Face / BERT)

Subword tokenization splits rare words into smaller, frequently used pieces.

```python
from transformers import AutoTokenizer

# Load the BERT uncased tokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

text = "unhappiness"

tokens = tokenizer.tokenize(text)



print("Subword Tokens:", tokens)
```

--- 

### 7. Byte Pair Encoding (BPE) Demonstration

BPE is a common subword algorithm. Here we see how the tokenizer handles a common word.

```python
tokens = tokenizer.tokenize("lower")

print("BPE Tokens:", tokens)
```

--- 

### 8. TensorFlow / Keras Tokenizer

This tokenizer maps words to unique numerical indices for deep learning models.

```python
from tensorflow.keras.preprocessing.text import Tokenizer

texts = ["I love learning NLP", "NLP is fun"]

tokenizer = Tokenizer()

# Builds the vocabulary index

tokenizer.fit_on_texts(texts)

print("Word Index:", tokenizer.word_index)
```

--- 

### 9. Gensim Tokenization

Gensim provides a simple and fast tokenization/cleaning method.

```python
from gensim.utils import simple_preprocess

text = "I love learning NLP!"

tokens = simple_preprocess(text)

print("Gensim Tokens:", tokens)
```

--- 

### 10. Moses Tokenizer (Sacremoses)

A tokenizer often used in machine translation for precise handling of punctuation and contractions.

```python
from sacremoses import MosesTokenizer

mt = MosesTokenizer()

tokens = mt.tokenize("I love learning NLP!")

print("Moses Tokens:", tokens)
```
